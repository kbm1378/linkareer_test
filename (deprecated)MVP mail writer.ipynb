```python
def benefit_translate(benefits):
    result_benefit = ""
    if("-" in str(benefits)):
        return "-"
    elif("," in str(benefits)):
        list_benefits = benefits.split(",")
    else :
        list_benefits = [str(benefits)]
    for b in list_benefits:
        if(b=="1"): result_benefit+= "활동비"
        elif(b=="2"): result_benefit+= "자사제품"
        elif(b=="3"): result_benefit+= "실무교육"
        elif(b=="4"): result_benefit+= "전문가/실무자 멘토링"
        elif(b=="5"): result_benefit+= "행사참여"
        elif(b=="6"): result_benefit+= "활동자간 교류"
        elif(b=="7"): result_benefit+= "수료증 및 인증서"
        elif(b=="8"): result_benefit+= "입사시 가산점"
        elif(b=="9"): result_benefit+= "임직원 만남"
        if("," in str(benefits)):
            if(benefits.index(b) != len(benefits)-1): result_benefit+=", "
    return result_benefit 

def email_to_user(email):
    return email.split("@")[0]
```
