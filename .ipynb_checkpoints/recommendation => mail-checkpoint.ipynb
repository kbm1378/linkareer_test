```python
file_name ="data/recommendation.xlsx"
xls = pd.ExcelFile(file_name)
dt_rec_per1 = xls.parse('Per1', header=1)
dt_rec_per2 = xls.parse('Per2', header=1)
dt_rec_per3 = xls.parse('Per3', header=1)

del dt_rec_per1[0]
del dt_rec_per2[0]
del dt_rec_per3[0]
```

```python
file_name ="data/activity_raw.xlsx"
xls = pd.ExcelFile(file_name)
dt_activity = xls.parse('Sheet1', header=0)
dt_activity = dt_activity.reindex(dt_activity.index.drop(0))  #"활동명", "마감일" 등 한글로 써져 있는거 지움
dt_activity = dt_activity.reindex(dt_activity.index.drop(1))  # 활동혜택의 세부사항때문에 생기는 두번째 줄도 지움

dt_activity.index = dt_activity['activity']
del dt_activity['activity']

dt_activity.head(1)
```

```python
def benefit_translate(benefits):
    result_benefit = ""
    if("-" in str(benefits)):
        return "-"
    elif("," in str(benefits)):
        list_benefits = benefits.split(", ")
    else :
        list_benefits = [str(benefits)]
    for b in list_benefits:
        if(b=="b1"): result_benefit+= "활동비"
        elif(b=="b2"): result_benefit+= "자사제품"
        elif(b=="b3"): result_benefit+= "실무교육"
        elif(b=="b4"): result_benefit+= "전문가/실무자 멘토링"
        elif(b=="b5"): result_benefit+= "행사참여"
        elif(b=="b6"): result_benefit+= "활동자간 교류"
        elif(b=="b7"): result_benefit+= "수료증 및 인증서"
        elif(b=="b8"): result_benefit+= "입사시 가산점"
        elif(b=="b9"): result_benefit+= "임직원 만남"
        if("," in str(benefits)):
            if(list_benefits.index(b) != len(list_benefits)-1): result_benefit+=", "
    return result_benefit 

def email_to_user(email):
    return email.split("@")[0]

def get_cafe_url(data):
    if(data=="-") : return "-"
    else : return dt_activity['url'][data]
        
def get_img_url(data):
    if(data=="-") : return "-"
    else : return dt_activity['img'][data]
```

```python
def write_a_benefit(i, user, dt_rec):
    if(benefit_translate(dt_rec['r'+str(i)+'_b'][user])=="-"):
        return ""
    else :
        html = """선호하는 혜택인 <span style="font-weight : bold;">
"""+benefit_translate(dt_rec['r'+str(i)+'_b'][user])+"""
      </span>의 혜택이 제공됩니다."""
    return html

def write_a_interest(i, user, dt_rec):
    if(dt_rec['r'+str(i)+'_i'][user]=="-"):
        return ""
    else :
        html = """관심 분야인 <span style="font-weight : bold;">
"""+dt_rec['r'+str(i)+'_i'][user]+"""
       </span> 관련 활동입니다.<br>"""
    return html

def write_a_industry(i, user, dt_rec):
    if(dt_rec['r'+str(i)+'_d'][user]=="-"):
        return ""
    else :
        html = """관심 산업군인 <span style="font-weight : bold;">
"""+dt_rec['r'+str(i)+'_d'][user]+"""
       </span> 관련 활동입니다.<br>"""
    return html

def write_a_score(i, user, dt_rec):
    score_ratio = round(5* dt_rec['r'+str(i)+'_s'][user]/dt_rec['r1_s'][user], 1)
    if score_ratio == 5.0 : star = "&#9733;&#9733;&#9733;&#9733;&#9733;"
    elif score_ratio >= 4.0 : star = "&#9733;&#9733;&#9733;&#9733;&#9734;"
    elif score_ratio >= 3.0 : star = "&#9733;&#9733;&#9733;&#9734;&#9734;"
    elif score_ratio >= 2.0 : star = "&#9733;&#9733;&#9734;&#9734;&#9734;"
    elif score_ratio >= 1.0 : star = "&#9733;&#9734;&#9734;&#9734;&#9734;"
    else : star = "&#9734;&#9734;&#9734;&#9734;&#9734;"
    return star + " ("+str(score_ratio)+" 점)"
    
    
def write_a_rec(i, user, dt_rec):
    if(dt_rec['r'+str(i)][user]=="-"): 
        if(i==1) : print "추천 전혀 없는 사용자 : ", dt_rec['email'][user]
        return ""
    else:
        html = "<!--"+str(i)*20+""" -->
      <table class="layout layout--no-gutter" style="border-collapse: collapse;table-layout: fixed;Margin-left: auto;Margin-right: auto;overflow-wrap: break-word;word-wrap: break-word;word-break: break-word;background-color: #fefdfe;" align="center"><tbody><tr><td colspan="2" class="column" style="padding: 0;text-align: left; vertical-align: top; color: #8e8e8e;font-size: 14px;line-height: 21px;font-family: Avenir,sans-serif;width: 600px;"><div style="Margin-left: 20px;Margin-right: 20px;Margin-top: 24px;Margin-bottom: 0px;"><h2 class="size-22" style="Margin-top: 20px;Margin-bottom: 0;font-style: normal;font-weight: normal;color: #44596b; font-size: 20px;"><span style="color:#026ec3"><strong>
      """+str(i)+". "+dt_rec['r'+str(i)][user]+"""</strong></span>&nbsp;&nbsp;<a href="
"""+get_cafe_url(dt_rec['r'+str(i)][user])+"""
      " target="_blank" style="text-decoration:underline; color:#036EC3; font-size : 13px; ">자세히 알아보기 </a></h2></div></td></tr><tr><td class="column" style="padding: 0;text-align: left;vertical-align: top;color: #8e8e8e;font-size: 14px;line-height: 21px;font-family: Avenir,sans-serif;width: 300px;"><div style="Margin-left: 20px;Margin-right: 20px;Margin-top: 24px;Margin-bottom: 24px;"><div class="image" style="font-size: 12px;font-style: normal;font-weight: 400;" align="center"><img style="display: block;border: 0;max-width: 410px; border: 1px solid #645353;" src="
"""+str(get_img_url(dt_rec['r'+str(i)][user]))+"""
      " alt="" width="260" height="260"></div></div></td><td class="column" style="padding: 0;text-align: left;vertical-align: top;color: #8e8e8e;font-size: 14px;line-height: 21px;font-family: Avenir,sans-serif;width: 300px;"><div style="Margin-left: 20px; Margin-right: 20px; Margin-top:22px;"><p class="size-16" style="Margin-top: 16px;Margin-bottom: 0;font-size: 16px;line-height: 24px;"><font color="#645353"><span style="font-weight : bold;">추천 이유</span>: <br><span style="font-size : 13px;">
"""+write_a_interest(i,user, dt_rec)+write_a_industry(i,user, dt_rec)+write_a_benefit(i, user, dt_rec)+""" 
       </span><br><br><span style="font-weight : bold;">예상 만족도</span> :<br>"""+write_a_score(i, user, dt_rec)+"""
       </font> </p></div></td></tr></tbody></table>"""
    return html
```
